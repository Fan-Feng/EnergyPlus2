```python
import os
import re
import glob as gb
files = gb.glob("*.cc")
```

```python
re_setup = re.compile(r'SetupOutputVariable\(.*?,.*?,.*?,(.*?),.*?\);', re.DOTALL)
```

```python
types = set()

for fpath in files:
    with open(fpath, 'r') as f:
        content = f.read()
    for m in re_setup.finditer(content):
        t = m.groups()[0].strip().replace('"', '').upper()
        types.add(t)
        if t not in ["HVAC", "PLANT", "SYSTEM", "ZONE"]:
            print(fpath, m)
            print(content[m.start():m.end()])
```

```python
fpath = 'CondenserLoopTowers.cc'
with open(fpath, 'r') as f:
    content = f.read()
for m in re_setup.finditer(content):
    t = m.groups()[0].strip().replace('"', '').upper()
    types.add(t)
    if t not in ["HVAC", "PLANT", "SYSTEM", "ZONE"]:
        print(fpath, m)
        break
```
